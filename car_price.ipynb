```python
from google.colab import files
import pandas as pd
import seaborn as sns
cars_data = pd.read_csv('/content/gdrive/MyDrive/cars.csv')
brand_data = pd.read_csv('/content/gdrive/MyDrive/brand.csv')
```

```python
cars_data['new_title'] = cars_data['title'].str.split().str[0]
cars_data['new_title']
```

```python
merged_data = cars_data.merge(brand_data, left_on='new_title', right_on='title_upper', how='left')
merged_data
```

```python
merged_data.drop(['title_y', 'new_title'], axis=1, inplace=True)
```

```python
merged_data
```

```python
# 중복된 행만 있는 데이터프레임 생성
duplicate_rows = merged_data[merged_data.duplicated(keep=False)]
duplicate_rows
```

```python
merged_data = merged_data.drop(index = duplicate_rows.index)  #중복행삭제
```

```python
#뽑은 중복행끼리 그룹화
grouped_data = duplicate_rows.groupby(duplicate_rows.columns.tolist(),dropna=False)
```

```python
#그룹화 결과 count
grouped_counts = grouped_data.size().reset_index(name='count')
grouped_counts
```

```python
grouped_counts['count'].sum() #그룹 count결과 검정
```

```python
# 그룹화된 결과를 출력
for name, group in grouped_data:
    count = len(group)  # 각 그룹의 행 개수
    print(f"Count: {count}, Group: {name}")
    print(group.to_string(index=False))
    print()  # 각 그룹 사이에 빈 줄 추가
```

```python
#중복값 없는 데이터 + 중복 count된 unique 데이터
combined_data = pd.concat([merged_data, grouped_counts], ignore_index=True)
```

```python
combined_data['count'] = combined_data['count'].fillna(1)
combined_data['count'].sum()
```
